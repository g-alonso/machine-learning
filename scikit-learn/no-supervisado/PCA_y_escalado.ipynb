```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
#%matplotlib inline

from pandas.plotting import scatter_matrix

wine = load_wine()
df_wine = pd.DataFrame(wine['data'], columns=[wine['feature_names']])
print("graficamos la data")
scatter_matrix(df_wine.iloc[:,0:4], alpha=0.2, figsize=(6, 6), diagonal='kde');

# --------- PCA SIN ESCALAR-----
pca = PCA(n_components=2)
pca_model=pca.fit(df_wine)
df_trans=pd.DataFrame(pca_model.transform(df_wine), columns=['pca1', 'pca2'])
plt.figure(0)
plt.scatter(df_trans['pca1'], df_trans['pca2'], alpha=0.8, c=wine["target"])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
# ------------------------------

# -------- ESCALAMOS -----------
df = StandardScaler().fit_transform(df_wine)
df = pd.DataFrame(df,columns=wine.feature_names)
print("graficamos la data")
scatter_matrix(df.iloc[:,0:4], alpha=0.2, figsize=(6, 6), diagonal='kde');
# ------------------------------

# --------- PCA ESCALADO--------
pca = PCA(n_components=2)
pca_model=pca.fit(df)
df_trans=pd.DataFrame(pca_model.transform(df), columns=['pca1', 'pca2'])
plt.figure(2)
plt.scatter(df_trans['pca1'], df_trans['pca2'], alpha=0.8, c=wine["target"])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
# ------------------------------
```
